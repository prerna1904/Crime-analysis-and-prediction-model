```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print("Libraries loaded successfully!")
```

```python
df = pd.read_csv('../data/crime_data.csv')

print(f"Dataset Shape: {df.shape}")
print(f"\nColumn Names:\n{df.columns.tolist()}")
print(f"\nFirst 5 rows:")
df.head()
```

```python
print("Dataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nBasic Statistics:")
df.describe()
```

```python
# Keep only the columns we need
columns_to_keep = [
    'Date', 'Primary Type', 'Description', 'Location Description',
    'Arrest', 'Domestic', 'Beat', 'District', 'Ward',
    'Community Area', 'Latitude', 'Longitude', 'Year'
]

columns_to_keep = [col for col in columns_to_keep if col in df.columns]
df = df[columns_to_keep]

print(f"Shape after dropping columns: {df.shape}")
df.head()
```

```python
# Drop rows where Latitude/Longitude is missing
df = df.dropna(subset=['Latitude', 'Longitude'])

# Fill missing categorical columns with 'Unknown'
for col in ['Location Description']:
    df[col] = df[col].fillna('Unknown')

# Fill missing District/Ward with their median
df['District'] = df['District'].fillna(df['District'].median())
df['Ward'] = df['Ward'].fillna(df['Ward'].median())

print("Missing values after cleaning:")
print(df.isnull().sum())
print(f"\nShape after cleaning: {df.shape}")
```

```python
# Fix remaining Community Area missing values
df['Community Area'] = df['Community Area'].fillna(df['Community Area'].median())

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')

# Extract time features
df['Hour']      = df['Date'].dt.hour
df['Day']       = df['Date'].dt.day
df['Month']     = df['Date'].dt.month
df['DayOfWeek'] = df['Date'].dt.dayofweek
df['DayName']   = df['Date'].dt.day_name()
df['MonthName'] = df['Date'].dt.month_name()
df['Season']    = df['Month'].apply(lambda x: 
                    'Winter' if x in [12,1,2] else
                    'Spring' if x in [3,4,5] else
                    'Summer' if x in [6,7,8] else 'Fall')

# Convert Arrest and Domestic to integer
df['Arrest']   = df['Arrest'].astype(int)
df['Domestic'] = df['Domestic'].astype(int)

print("Date features extracted successfully!")
print(df[['Date','Hour','Day','Month','DayOfWeek','DayName','Season']].head())
```

```python
def categorize_crime(crime_type):
    violent      = ['HOMICIDE', 'ASSAULT', 'BATTERY', 'ROBBERY', 
                    'CRIM SEXUAL ASSAULT', 'SEX OFFENSE', 'KIDNAPPING']
    property_crime = ['THEFT', 'BURGLARY', 'MOTOR VEHICLE THEFT', 
                      'ARSON', 'CRIMINAL DAMAGE']
    drug         = ['NARCOTICS', 'OTHER NARCOTIC VIOLATION', 'DRUG']
    fraud        = ['FRAUD', 'FORGERY', 'COUNTERFEITING', 'IDENTITY THEFT']
    public_order = ['PUBLIC PEACE VIOLATION', 'WEAPONS VIOLATION',
                    'LIQUOR LAW VIOLATION', 'GAMBLING', 'DISORDERLY CONDUCT']
    
    crime_type = str(crime_type).upper()
    if any(c in crime_type for c in violent):          return 'Violent'
    elif any(c in crime_type for c in property_crime): return 'Property'
    elif any(c in crime_type for c in drug):           return 'Drug'
    elif any(c in crime_type for c in fraud):          return 'Fraud'
    elif any(c in crime_type for c in public_order):   return 'Public Order'
    else:                                              return 'Other'

df['Crime Category'] = df['Primary Type'].apply(categorize_crime)

print("Crime Category Distribution:")
print(df['Crime Category'].value_counts())
```

```python
df.to_csv('../data/crime_data_cleaned.csv', index=False)
print(f"Cleaned data saved!")
print(f"Final shape: {df.shape}")
print(f"\nColumns in cleaned data:")
print(df.columns.tolist())
```
